```python
#Import pandas, numpy, seaborn,matplotlib
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the data in a dataframe
df = pd.read_csv('ie_cpi_data.csv')
```

```python
X=df['Date']
```

```python
df = pd.DataFrame(X, columns=["Date","Goods","Services","Utilities and Local Charges","Energy Products"])
```

```python
df.plot()
```

```python
x = 'Date'
sns.countplot(data=df, x=df[x])
plt.show()

#heatMap using seaborn library
sns.heatmap(df.corr(), cmap='coolwarm')

#seaborn color palette setting
sns.color_palette(palette=None, n_colors=None, desat=None, as_cmap=False)
plt.title('titanic.corr()')
plt.show()
```
